## Bubble Sort

```python
import random
```

```python
arreglo = []
for i in range(10000):
  arreglo.append(random.randint(1, 100000))
```

```python
flag = False
while flag==False:
  banda = True
  for i in range(len(arreglo)-1):
    if arreglo[i]<arreglo[i+1]:
      aux = arreglo[i]
      arreglo[i] = arreglo[i+1]
      arreglo[i+1] = aux
      flag = False
print(arreglo)
```
