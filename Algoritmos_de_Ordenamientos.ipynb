# Algoritmos de Ordenamiento

## Bubble Sort

```python
import random
```

```python
def bubbleSort1(arreglo):
  flag = False
  while flag==False:
    flag = True
    for i in range(len(arreglo)-1):
      if arreglo[i]>arreglo[i+1]:
        aux = arreglo[i]
        arreglo[i] = arreglo[i+1]
        arreglo[i+1] = aux
        flag = False
  print(arreglo)
```

Otra manera de desarrollarlo

```python
def bubbleSort2(arreglo):
  n = len(arreglo)
  for i in range(n):
    for j in range(n-1):
      if arreglo[j] > arreglo[j+1]:
        aux = arreglo[j]
        arreglo[j] = arreglo[j+1]
        arreglo[j+1] = aux
  print(arreglo)
```

```python
arr = []
for i in range(10):
  arr.append(random.randint(1, 1000))
bubbleSort1(arr)
bubbleSort2(arr)
```
